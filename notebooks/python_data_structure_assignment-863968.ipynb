
### üìù Task 1: Shopping Cart System (List)

Write a Python program that simulates a simple shopping cart:

- Store product names in a list
- Allow the user to:
  - Add new products to the cart
  - Remove a product from the cart
  - View all products in the cart
- Display the total number of items in the cart

```python
"""
Shopping Cart System

This program simulates a simple shopping cart. Users can add or remove products, view all the items in the cart,
see the total number of items or exit the cart.
Menu options:
1. Add product
2. Remove product
3. View Cart
4.Total items in cart
5. Exit

This program will continue running until user chooses to exit
"""

#empty list
cart=[]
#Menu options to choose from
while True:
    print("\n Shopping cart menu")
    print("1. Add Product")
    print("2. Remove Product")
    print("3. View Cart")
    print("4. Total Items in Cart")
    print("5. Exit")

    choice=input("Choose an option:")
     
     #option1
    if choice=="1":
        product=input("Enter product name to add")
        cart.append(product)
        print(f"{product} added to cart")

#option2
    elif choice=="2":
        product=input("Enter product to be removed")
        if product in cart:
            cart.remove(product)
            print(f"\n{product} removed from cart")
        else:
            print(f"\n{product} not found in cart")


    elif choice=="3":
      if cart:
        print("\nProducts in your cart:")
        for items in cart:
            print(f"-{items}")
      else:
          print("Cart is empty")


    elif choice=="4":
        print(f"Total items in cart {len(cart)}")


    elif choice=="5":
        print("Thank you for shopping with us")
        break

    else:
        print("Invalid choice. Choose between 1-5")



```

### üìù Task 2: Student Attendance System (Set)

Write a Python program that manages student attendance:

- Store roll numbers of students present today in a set
- Store roll numbers of students present yesterday in another set
- Display:
  - Students present on both days
  - Students present only today
  - Students absent today but present yesterday


```python
"""
This program is to store roll numbers of students who were present today in a set and those of yesterday in another set.
Displays students presents on both day
Displays students present only today
Displays students present only yesterday but absent today
"""


#set for today's attendance
today={101,102,103,104,105}

#set for yesterday's
yesterday={101,105,107,109,110}
#& to show students present on both days
both_days =  today & yesterday

#to find student who came only today
only_today = today - yesterday

#to find student who came only yesterday
absent_today = yesterday  -  today

#displaying results
print("Students present on both days:", both_days)
print("Student present only today:", only_today)
print("Student absent today but present yesterday:", absent_today)
```

Write a Python program that represents bus route information:

- Store bus route details (Route Number, Start Point, End Point, Distance) in a tuple

- Explain why tuple is suitable for this data



```python
"""
Bus Route Information Program

This program stores bus route details such as route number,
start point, end point, and distance using a tuple.

A tuple is used because bus route information is permanent
and should not be modified. Tuples are immutable, making
them suitable for fixed data like route details.
"""

#Tuple for Bus info
Bus_route=(101, "Baneshwor", "Dillibazar", "3KM")

#printing details
print("Bus Route Details")
print("Route Number:", Bus_route[0])
print("Start Point:", Bus_route[1])
print("End Point", Bus_route[2])
print("Distance:", Bus_route[3])


#Tuple is suitable for this data because bus route details are permanent and tuple is a data structure which is immutable.
```

### üìù Task 4: Exam Marks Analysis (List + Tuple)

Write a Python program that:

- Stores marks of students in a list
- Calculates:
  - Highest mark
  - Lowest mark
  - Average mark
- Stores the calculated results in a tuple
- Displays all values clearly


```python
"""
Student Marks Analysis Program

This program accepts marks of multiple students and stores them in a list.
It calculates the highest mark, lowest mark, and average mark using
built-in functions.

The calculated results are stored in a tuple since the final results
(highest, lowest, and average) should not be modified.
"""


#no of students
n=int(input("Enter the number of students"))

#empty list
marks=[]

#for loop to insert marks of student
for i in range(n):
    m=int(input(f"Enter the marks of subject {i+1}"))
    marks.append(m)

#to find maximum marks
highest=max(marks)

#to find minimum marks
lowest=min(marks)


#to find average marks
Average=sum(marks)/n

#storing the result in tuple
results=(highest, lowest, Average)

#displaying final result
print("Marks list", marks)
print("Highest Mark:", results[0])
print("Lowest Mark:", results[1])
print("Average Mark:", results[2])
```

### üìù Task 5: Duplicate Data Cleaner (List ‚Üí Set)

Write a Python program that:

- Takes a list of email addresses as input
- Removes duplicate email addresses using a set
- Converts the cleaned data back into a list
- Displays both original and cleaned lists


```python
"""
Email List Deduplication Program

This program accepts a list of email addresses from the user.
It converts the list to a set to remove duplicate emails, 
and then converts it back to a list to display the cleaned data.

- Original list shows all emails including duplicates.
- Cleaned list contains only unique email addresses.
"""

#to take input for numbers of email
n=int(input("Enter the number of emails"))


#creating list for email
email_list=[]

#for loop to insert email
for i in range(n):
    mail=input((f"Enter email {i+1}"))
    email_list.append(mail)

#converting list to set
email_set=set(email_list)

#converting the cleaned set to list
email_list1=list(email_set)

#printing both original and cleaned list
print("Original list:", email_list)

print("Cleaned list", email_list1)

```
