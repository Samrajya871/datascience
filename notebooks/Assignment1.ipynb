### üìù Task 2: ATM Withdrawal System

Write a Python program that:

- Initializes an account balance of **NPR 25,000**  
- Takes the withdrawal amount as input from the user  
- Allows withdrawal only if:
  - The withdrawal amount is a multiple of **500**
  - The remaining balance is not less than **NPR 1,000**
- Displays appropriate messages for successful or failed transactions


```python
account_balance=25000
withdrawal=int(input("Enter the amount to be withdrawn"))
if withdrawal%500!=0 :
    print("Withdraw amount must be multiple of 500")
elif account_balance-withdrawal<1000:
    print("Transaction Failed: Account Balance should be minimum 1000")
elif withdrawal>account_balance:
    print("Insufficient Balance")
else:
  remaining_balance=account_balance-withdrawal    
  print("Amount has been withdrawn")
  print("Remaining Balance =",remaining_balance)
    
    
      
```

### üìù Task 4: Student Grade Evaluation Function

Write a Python function that:

- Accepts marks of **five subjects** as input  
- Calculates the average marks  
- Returns the grade based on the average score

```python
sub1 = int(input("Enter marks of 1st Subject: "))
sub2 = int(input("Enter marks of 2nd Subject: "))
sub3 = int(input("Enter marks of 3rd Subject: "))
sub4 = int(input("Enter marks of 4th Subject: "))
sub5 = int(input("Enter marks of 5th Subject: "))

averagemarks = (sub1 + sub2 + sub3 + sub4 + sub5) / 5
print("Average Marks =", averagemarks)

if averagemarks >= 90:
    print("Grade: A+")
elif averagemarks >= 80:
    print("Grade: A")
elif averagemarks >= 70:
    print("Grade: B+")
elif averagemarks >= 60:
    print("Grade: B")
elif averagemarks >= 50:
    print("Grade: C")
elif averagemarks >= 40:
    print("Grade: D")
else:
    print("Grade: Fail")
```

### üìù Task 5: Menu-Driven Calculator Using Functions

Write a Python program that:

- Uses separate functions for:
  - Addition
  - Subtraction
  - Multiplication
  - Division
- Performs the selected operation



```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    if b==0:
        return "error: Divison by zero"
    return a/b
num1=float(input("Enter the first number"))
num2=float(input("Enter the second number"))
operation=input("Enter operation: +, -, *, /")

if operation == "+":
    result=add(num1,num2)
elif operation == "-":
    result=sub(num1,num2)
elif operation == "*":
    result=mul(num1,num2)
elif operation == "/":
    result=div(num1,num2)
else:
    result="invalid output"

print("Result=", result)
       
  
 


    

    
```

Using **lambda functions**, write programs to:

- Check whether a number is even or odd  
- Find the maximum of two numbers

```python
odd_even=lambda x:"Even" if x%2==0 else "odd"
print(odd_even(4))
print(odd_even(7))
    
```

```python
maximum=lambda x,y:x if x>y else y
print(maximum(4,5))
```

```python
cube=lambda x: x*x*x
print(cube(10))
```

### üìù Task 8: Banking System Error Handling

Write a Python function `withdraw()` that:

- Accepts balance and withdrawal amount as input  
- Raises an error if the withdrawal amount is greater than the balance  
- Handles invalid inputs using `try-except` blocks


```python
def balance_withdraw(balance, withdraw):
    if withdraw > balance:
        raise ValueError("Insufficient Funds")
    return balance - withdraw


try:
    balance = int(input("Enter the balance of account: "))
    withdraw = int(input("Enter the amount to withdraw: "))
    remaining = balance_withdraw(balance, withdraw)
    print("Remaining balance:", remaining)

except ValueError as e:
    print("Error:", "Invalid input")
```
